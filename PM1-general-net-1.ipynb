```python
import numpy
import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx
```

Aarchivo excell:  PM1-content-task-net-2024   hoja:  General

```python
archivo_nombre="PM1-content-task-net-2024.xlsx"
archivo=pd.read_excel(archivo_nombre, sheet_name="General")
print(archivo)
```

Generar la gráfica

```python
G=nx.from_pandas_edgelist(archivo,"IN","OUT")
nx.draw_kamada_kawai(G,node_color='b',node_size=300,alpha=0.6,with_labels=True)
plt.show()
```

```python
G=nx.from_pandas_edgelist(archivo,"IN","OUT")
nx.draw_random(G,node_color='b',node_size=300,alpha=0.6,with_labels=True)
plt.show()
```

```python
G=nx.from_pandas_edgelist(archivo,"IN","OUT")
nx.draw_spring(G,node_color='b',node_size=300,alpha=0.6,with_labels=True)
plt.show()
```
