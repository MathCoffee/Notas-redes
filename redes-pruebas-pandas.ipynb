Cargar librerías


```python
import numpy
import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx
```

Cargar archivo de excell

```python
archivo_nombre="PANDAS-first-net.xlsx"
archivo=pd.read_excel(archivo_nombre, sheet_name="Aristas")
print(archivo)


```

Generar la gráfica

Gráfica circular

```python
G=nx.from_pandas_edgelist(archivo,"Entrada","Salida")
nx.draw_circular(G,node_color='b',node_size=500,alpha=0.3,with_labels=True)
plt.show()
```

Gráfica Kamada Kawaii

```python
G=nx.from_pandas_edgelist(archivo,"Entrada","Salida")
nx.draw_kamada_kawai(G,node_color='b',node_size=500,alpha=0.3,with_labels=True)
plt.show()
```

Gráfica draw_spectral


```python
G=nx.from_pandas_edgelist(archivo,"Entrada","Salida")
nx.draw_spectral(G,node_color='b',node_size=600,alpha=0.5,with_labels=True)
plt.show()
```
