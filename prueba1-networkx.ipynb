Cargando Librerias

```python
import numpy
import pandas as pd 
import matplotlib.pyplot as plt
import networkx as nx
```

Gr√°ficas no dirigidas

```python
G = nx.Graph()
G.add_nodes_from([2,'x','y',9])
G.add_edges_from([(2,'x'),('x','y'),('y',9)])
nx.draw(G)
plt.show()	
```

N√∫mero de v√©rtices y aristas de una gr√°fica

```python
G.order()
G.size()   
print(G.order())
print(G.size())
```

Vista de nodos

```python
G.nodes()		

```

Vista de aristas

```python
G.edges()	
```

Longitud de la gr√°fica

```python
print(len(G))
```

Modificaciones a la gr√°fica

```python
G.add_node(12)				#agrega un nodo
nx.draw(G)
plt.show()	
G.add_edge(23,45)			#agrega aristas, incluso si no tiene nodos
nx.draw(G)
plt.show()	
G.remove_node(9)				#elimina nodos
nx.draw(G)
plt.show()	
G.remove_edges_from( [(2,'x'),(23,45)] ) 
nx.draw(G)
plt.show()	
```
