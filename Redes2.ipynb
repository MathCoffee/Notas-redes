Cargar librerías

```python
import numpy
import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx
```

Gráfica dirigida 1


```python
G = nx.Graph()
G.add_nodes_from([2,'x','y',9])
G.add_edges_from([(2,'x'),('x','y'),('y',9)])
nx.draw(G)
plt.show()
G.order()
G.size()      
```

```python
G.nodes()				#muestra los nodos
G.edges()	
```

Se agrega un nodo a la gráfica y se calcula el orden

```python
G.add_node(12)	
nx.draw(G)
plt.show()
G.order()
G.size()      
```

Se agrega una arista a la gráfica y se calcula el orden


```python
G.add_edge('y',12)
nx.draw(G)
plt.show()
G.order()
G.size()   
```

Se agrega una arista incluso si los nodos no estaban antes.  Se calcula orden

```python
G.add_edge(23,45)			#agrega aristas, incluso si no tiene nodos
nx.draw(G)
plt.show()
G.order()
G.size()   
```

Para hacer la gráfica conexa se agrega una arista nueva que conecte. 

```python
G.add_edge(23, 'x')
nx.draw(G)
plt.show()
G.order()
G.size()   
```

Agregamos dos aristas más para generar dos ciclos. Y se calcula el orden
